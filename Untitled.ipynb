```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('shopping_trends.csv')
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
numerical_col = ['Age', 'Purchase Amount (USD)','Review Rating','Previous Purchases']
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data[numerical_col].corr(), annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='Gender', data=data,palette='viridis')
plt.title('Gender Disatribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='Item Purchased', data=data, palette='pastel', order=data['Item Purchased'].value_counts().index)
plt.title('Most Purchased Items')
plt.xlabel('Count')
plt.ylabel('Item Purchased')
plt.show()
```

```python
data.columns
```

```python
sns.scatterplot(x='Age',y='Purchase Amount (USD)', hue='Category', data=data, palette='cool')
plt.title('Age Vs Purchased Amount')
plt.show()
```

```python
sns.boxplot(x='Frequency of Purchases', y='Review Rating', data=data,palette='Set2')
plt.title('Review Rating Across Purchase Frequency')
plt.show()
```

```python
sns.pairplot(data[numerical_col], diag_kind='kde', plot_kws={'alpha':0.7}, diag_kws={'shade':True})
plt.show()
```
